### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
````

Playstore app dataset under Size column present many String and object than i will be convert object to float

````python
Playstore_df['Size'] = Playstore_df.Size.apply(lambda x: x.strip('+'))# Removing the + Sign
````

````python
Playstore_df['Size'] = Playstore_df.Size.apply(lambda x: x.replace(',', ''))# For removing the `,`
````

````python
Playstore_df['Size'] = Playstore_df.Size.apply(lambda x: x.replace('M', 'e+6'))# For converting the M to Megabytes
````

````python
Playstore_df['Size'] = Playstore_df.Size.apply(lambda x: x.replace('k', 'e+3'))# For converting the K to Kilobytes
````

````python
Playstore_df['Size'] = Playstore_df.Size.replace('Varies with device', np.NaN)
````

````python
Playstore_df['Size'] = pd.to_numeric(Playstore_df['Size']) # Converting the string to Numeric type
````

````python
def findtop10incategory(str):
    str = str.upper()
    top10_category = Google_Playstore_df[Google_Playstore_df['Category'] == str]
    top10_Installs_app = top10_category.sort_values(by='Installs', ascending=False).head(10)
    return top10_Installs_app
````

````python
top10apps = findtop10incategory('FAMILY')
````
